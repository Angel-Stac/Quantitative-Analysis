```python
#Initializing variables and datasets
x = [0.18, 1.0, 0.92, 0.07, 0.85, 0.99, 0.87]
y = [109.85, 155.72, 137.66, 76.17, 139.75, 162.6, 151.77]
w = 0
c = 0
L = 0.001
Dw =[]
Dc = []
lenofx  = len(x)
it = 2000
```

```python
max_length = max(len(x), len(y))
```

```python
#initializing loop to compute using the gradient descent algorithm given ,
for i in range (it) : 
    Dw_list = []  # List to store individual Dw values
    Dc_list = []  # List to store individual Dc values
   # Compute predictions
    y_predictions = [x[j] * (w + c) for j in range(lenofx)]
    
    # Compute gradients
    for j in range(lenofx):
     Dw_list.append(x[j] * (y[j] - y_predictions[j]))
     Dc_list.append((y[j] - y_predictions[j]) )
    Dw = sum(Dw_list) / lenofx
    Dc = sum(Dc_list) / lenofx
    w -= L * Dw
    c -= L * Dc
    print("w = ",w, "c = ",c)





```

```python
# Given lists
x = [True, True, False, False, True, False, True]
y = [False, True, True, True, False, False, True]

# Element-wise and, or, not operations
result_and = [x[i] and y[i] for i in range(len(x))]  
result_or = [x[i] or y[i] for i in range(len(x))]    
result_not_x = [not x[i] for i in range(len(x))]     
result_not_y = [not y[i] for i in range(len(y))]    

# Print the results
print("Element-wise AND:", result_and)
print("Element-wise OR:", result_or)
print("Element-wise NOT x:", result_not_x)
print("Element-wise NOT y:", result_not_y)
```

```python
def is_true_40_percent_or_more(bool_list): #parameter will be an input list 
    # Calculate the total number of True elements
    total = sum(bool_list)  
    # Check if True is 40% or more
    return total >= 0.4 * len(bool_list)


list1 = [True, False, True, True, False, False, True]
result = is_true_40_percent_or_more(list1)
print("Result:", result)
```
